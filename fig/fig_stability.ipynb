```python
import os
import torch

os.environ["CUDA_VISIBLE_DEVICES"] = "0"
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model_name = 'opt-6.7b' #'llama3-8b'
checkpoint_path = '../models/opt-6.7b' #'../models/llama3-8b'
data_path = 'fancyzhx/ag_news'
```

# Step1：Load Model and Data

```python
from transformers import OPTForCausalLM,AutoModelForCausalLM
from transformers import AutoTokenizer

model = AutoModelForCausalLM.from_pretrained(checkpoint_path, device_map=device, torch_dtype=torch.float16)
tokenizer = AutoTokenizer.from_pretrained(checkpoint_path, device_map=device)
```

```python
from datasets import load_dataset
import os
import subprocess

dataset = load_dataset("allenai/c4",data_files={"train": "en/c4-train.00000-of-01024.json.gz"},
                        split="train")

print("Dataset Info:")
print(dataset)
```

```python
# Select a sentence from the C4 dataset for proof. You can replace it with any other sentence.
sentence=dataset['text'][17]
words = sentence.split()
data= [' '.join(words[:i+1]) for i in range(len(words))]
len(data)
```

# Step2：Collect Activation

```python
import torch
import re

activations = []

def get_activation(name, activation_dict):
    def hook(model, input, output):
        activation_dict[name] = input[0].detach().cpu()
    return hook

def register_act_hooks(model, model_name, activation_dict):
    hooks = []
    for name, layer in model.named_modules():
        if "opt" in model_name:
            if isinstance(layer, torch.nn.ReLU):
                hooks.append(layer.register_forward_hook(get_activation(name, activation_dict)))
            
        elif "llama" in model_name:
            if isinstance(layer, torch.nn.SiLU):
                hooks.append(layer.register_forward_hook(get_activation(name, activation_dict)))
                
    return hooks

def remove_hooks(hooks):
    for hook in hooks:
        hook.remove()

# hook activation
for i in range(len(data)):
    prompt = data[i]
    tokenized_input = tokenizer(prompt, return_tensors="pt").to(device)
    
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s.,!?;:]', '', prompt)
    tokenized_input = tokenizer(cleaned_text, return_tensors="pt", max_length=256, truncation=True).to(device)
    
    activation_dict = {}

    hooks = register_act_hooks(model, model_name, activation_dict)

    with torch.no_grad():
        outputs = model(**tokenized_input)
        
    activations.append(activation_dict)

    remove_hooks(hooks)
```

```python
# Choose any layer
Layer_num = 25
layer_act=[]
sentence_lenth=[]
for i in range(len(activations)):
    if "opt" in model_name:
        str_act="model.decoder.layers."+str(Layer_num)+".activation_fn"
    elif "llama" in model_name:
        str_act="model.layers."+str(Layer_num)+".mlp.act_fn"
        
    tensor = activations[i][str_act].cpu()
    layer_act.append(tensor.squeeze(0))
    if "opt" in model_name:
        m=tensor.size(0)
    elif "llama" in model_name:
        m=tensor.size(1)
    sentence_lenth.append(m)
A_tensor = torch.cat(layer_act, dim=0)
```

# Step3：Calculate Sentence-wise Core Neurons

```python
if "opt" in model_name:
    neuron_num = model.model.decoder.layers[0].fc1.weight.size(0)
elif "llama" in model_name:
    neuron_num = model.model.layers[0].mlp.down_proj.weight.size(1)
```

```python
sentence=[]
num=0
# sentence-wise and token-wise sparisity
sparisity = 0.2
token_sparisity = 0.2

for i in range(len(sentence_lenth)):
    lenth=sentence_lenth[i]
    list_now=list(range(num, num+lenth))
    sentence.append(list_now)
    num=num+lenth

act_all=(A_tensor).cpu()
count_act_all = (act_all > 0).sum(dim=1)
sorted_values, sorted_indices = torch.sort(act_all, dim=1, descending=True)

top50_indices=[]
for i in range(act_all.size(0)):
    indices = sorted_indices[i, :int(torch.round(count_act_all[i] * token_sparisity))]
    top50_indices.append(indices.tolist())

SEN_CORE=[]
for i in range(len(sentence)):
    cluster5= sentence[i]
    act_clu =  [top50_indices[i] for i in cluster5]
    data_flattened = [item for sublist in act_clu for item in sublist]
    data_flattened=torch.tensor(data_flattened)
    unique_numbers, counts = data_flattened.unique(return_counts=True, sorted=True)

    sorted_indices = torch.argsort(counts, descending=True)
    sorted_numbers = unique_numbers[sorted_indices]
    sorted_counts = counts[sorted_indices]
    neuron_remained = int(neuron_num * sparisity)
    SEN_CORE.append(sorted_numbers[:neuron_remained].numpy())
```

# Step4: Proof of stability through TSNE

```python
import numpy as np
import torch
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.manifold import TSNE

data = SEN_CORE


mlb = MultiLabelBinarizer(classes=range(0, 16383))
encoded_data = mlb.fit_transform(data)


tsne = TSNE(n_components=2,perplexity=20,random_state=0)
embedded = tsne.fit_transform(encoded_data)

fig, axs = plt.subplots(2, 1, figsize=(20, 10))
x = np.arange(0, 336)
cluster_points = embedded
axs[0].plot(x, cluster_points[:, 0], linewidth=4)
axs[0].set_xlabel('Num of Tokens', fontdict={'family': 'serif', 'size': 45})
axs[0].set_ylabel('Dimension 1', fontdict={'family': 'serif', 'size': 45})
axs[0].tick_params(labelsize=30)
axs[0].grid(True, linewidth=0.8, linestyle='-', alpha=1.0)
axs[0].set_xlim(0, 336)

axs[1].plot(x, cluster_points[:, 1], linewidth=4)
axs[1].set_xlabel('Num of Tokens', fontdict={'family': 'serif', 'size': 45})
axs[1].set_ylabel('Dimension 2', fontdict={'family': 'serif', 'size': 45})
axs[1].tick_params(labelsize=30)
axs[1].grid(True, linewidth=0.8, linestyle='-', alpha=1.0)
axs[1].set_xlim(0, 336)

plt.subplots_adjust(hspace=0.4)
plt.show()
```

# Step5: Stability Demonstrated by Heatmap

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec
from matplotlib.colors import ListedColormap

data_list=[]
list=[10,50,100,200,300]
for i in range(5):
    num=list[i]
    cluster5= sentence[num]
    act_clu =  [top50_indices[i] for i in cluster5]
    data_flattened = [item for sublist in act_clu for item in sublist]
    data_flattened=torch.tensor(data_flattened)
    unique_numbers, counts = data_flattened.unique(return_counts=True, sorted=True)
    number_counts = [0] * 256 #784 #3072 #16384
    
    for num, count in zip(unique_numbers, counts):
        if num<256:
            number_counts[num.item()] = count.item()

    number_counts_np = np.array(number_counts) / max(number_counts)
    
    heatmap_data = number_counts_np.reshape(16, 16)
    data_list.append(heatmap_data)
    
titles = ['length=10', 'length=50', 'length=100', 'length=200', 'length=300']


fig = plt.figure(figsize=(20, 5))
gs = gridspec.GridSpec(1, 6, width_ratios=[1, 1, 1, 1, 1, 0.05], wspace=0.3)

cmap = ListedColormap(['powderblue', 'deepskyblue','coral', 'tomato','orangered', 'red', 'firebrick','maroon'])

for i in range(5):
    ax = fig.add_subplot(gs[0, i])
    ax.set_xticks([])
    ax.set_yticks([])
    im = ax.imshow(data_list[i], cmap=cmap, interpolation='nearest')
    ax.set_title(titles[i], fontdict={'family':'serif','size': 20})
    # ax.axis('off')


cbar_ax = fig.add_subplot(gs[0, 5])
cbar = fig.colorbar(im, cax=cbar_ax)

cbar.set_label('Count', fontdict={'family':'serif','size': 20})
cbar.ax.tick_params(labelsize=15)

pos = cbar_ax.get_position()


new_height = pos.height * 0.7 
new_bottom = pos.y0 + (pos.height - new_height) / 2

new_pos = [pos.x0, new_bottom, pos.width, new_height]
cbar_ax.set_position(new_pos)


plt.tight_layout()
plt.savefig('hot_combine.pdf',bbox_inches='tight')
plt.show()
```

```python
import pickle
import sys
import os
sys.path.append("/".join(os.getcwd().split("/")[:-1]))
sys.path.append("/".join(os.getcwd().split("/")[:-2]))
import common
import torch
from sparsity_levels import *
from tqdm import tqdm

checkpoint_path = "/local/huzaifa/CoreInfer/models/llama3-3b"
dir_path = "/local/huzaifa/CoreInfer/sparsity_levels/llama3-3b/sparsity_levels_files"
model_name = "llama3-3b"
num_neurons = common.MODEL_INFO[model_name]["num_neurons"]
model, tokenizer, num_layers = common.load_model(model_name, 5, 27, checkpoint_path, "cpu", None)
```

```python
def process_dataset_levels(activations_path, goal_sparsity, dataset_name):
    with open(activations_path, "rb") as f:
        activations = pickle.load(f)
        
    possible_params, sparsity_levels = get_sparsity_levels_params(model_name, num_layers, activations, 1, 1000, goal_sparsity, 0.01, 5, 27)
    save_levels(sparsity_levels, dataset_name, dir_path + "/sparsity_levels10%")
    
    return possible_params, sparsity_levels
```
